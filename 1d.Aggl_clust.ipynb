```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
X=np.array([[1,	1],
           [1.5,	1.5],
           [5,	5],
           [3,	4],
           [4,  4],
           [3,	3.5]])
```

```python

# Generate sample data with 3 blobs
#X, _ = make_blobs(n_samples=500, centers=3, random_state=0)
plt.scatter(X[:, 0], X[:, 1])
```

```python
# Perform Agglomerative Hierarchical Clustering
agg_cluster = AgglomerativeClustering(n_clusters=2,  metric='euclidean', linkage='single')
labels = agg_cluster.fit_predict(X)
#clustering = AgglomerativeClustering().fit(X)
#clustering.labels_
agg_cluster.labels_
```

```python
# Plot the results

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Agglomerative Hierarchical Clustering')
plt.show()
```

```python
linkage_data = linkage(X, method='single', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```
